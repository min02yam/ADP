```r
require(dplyr)
require(magrittr)
require(ggplot2)
```

```r
install.packages("zoo")
library(zoo)
```

```r
#데이터 테이블 생성
dt = data.frame( times = as.character(seq(1:9)),
  value = c(22.9, 23.4, 21.1, 25.7, 27.4, 30.6, 31.8, 30.2, 41.1))
dt %>% head
```

```r
dt %<>% mutate(ma3 = zoo::rollmean(value , k = 3, fill = NA))
```

```r
dt %>%
 ggplot(aes(x=times)) +
 geom_line(aes(y=value, group = 1 , color = 'value')) +
 geom_line(aes(y=ma3, group = 1 , color = 'ma3')) +
 labs(y = "silver", x="month") +
 scale_color_manual(values = c('value' = 'red' , 'ma3' = 'blue'))
```
